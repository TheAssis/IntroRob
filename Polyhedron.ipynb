```
#https://stackoverflow.com/questions/39408794/python-3d-pyramid

from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d.art3d import Poly3DCollection, Line3DCollection
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

#Novas posições
w = [227.06306227150048, 0.0, 145.62076945605054]
a = [239.17423043680387,55.21772194479351,79.91174734242598]
s = [150.63160658264,-13.178557944625807,41.67013688186239]
d = [245.78825374991018,-56.74468954315267,58.18067787922098]

# vertices of a pyramid
v = np.array([w,s,d,a])
print(v.shape)


ax.scatter3D(v[:, 0], v[:, 1], v[:, 2])

# generate list of sides' polygons of our pyramid
verts = [ [v[0],v[1],v[2]], [v[1],v[2],v[3]],
 [v[0],v[2],v[3]], [v[0],v[1],v[3]] ]

# plot sides
ax.add_collection3d(Poly3DCollection(verts, 
 facecolors='green', linewidths=1, edgecolors='blue', alpha=.25))

plt.show()

print("Vértice topo: ", w)
print("Vértices da base: ",a,"\n",s,"\n",d )
```
